#### This is a combination of the previous lab and the previous homework. Create a SageMath interact that allows to user to specify a function, an interval [a,b] and an error of approximation. Then also add two checkboxes, one for Bisection Method, one for Newton-Raphson's Method. Then, your interact should:

- always display/plot the function f.
- Please use color red for this graph. 
- if the first checkbox is selected, it should also animate the successive approximation obtained using Bisection Method (this  is essentially what I wanted for the previous lab - we also did this in class)
- Please use color green for this graph. 
- if the second checkbox is selected, it should also animate the successive approximation obtained using Newton-Raphson's Method (this is the new part, bus similar to the the previous lab)
- Please use color blue for this graph. 

```python
@interact
def newton_rhaphson(f = input_box(default = x^2 - 2),
                   a = input_box(default = -1),
                   b = input_box(default = 3),
                   x0 = input_box(type = float, default = 2.0), 
                   max_steps = input_box(type = int, default = 10), 
                   max_error = input_box(default = 0.00001),
                   error = input_box(default = 0.00001),
                   newton = ("Newton-Raphson", False), 
                   bisection = ("Bisection Method", False), 
                   delay = slider(0,350,5,default=200,label="Set delay: ")):
    
    # plot function f(x)
    pl = plot(f(x), x, a, b,xmin=a-0.1,xmax=b+0.1, color='red')
    
    # point (x,y) values
    list_approx = [(x0, 0)]

    # plots 
    list_plots = [pl]
    
    # check for Newton-Raphson
    if newton == True:

        # function f(x)
        f(x) = f

        # function f'(x)
        f_prime(x) = diff(f(x), x)

        # keep track 
        step = 0

        # iterate to find x
        for step in range(max_steps):
            
            # check if solution
            if f(x0) == 0:
                print(f"exact solution is {x0.n(7)}")
                return

            # check if f'(x) is 0
            if f_prime(x0) == 0:
                print(f"newton's method won't work on critical points")
                break

            # guess
            x1 = x0 - f(x0)/f_prime(x0)

            # add to points values
            list_approx += [(x1, 0)]
            
            # print each time step
            print(f"{step}: x1 = {x1}, x0 = {x0}, f(x) = {f(x0)}, f'(x) = {f_prime(x0)}, |x1-x0| = {abs(x1-x0)}")

            # check for convergence
            if abs(x1-x0) < max_error:
                print(f"the approximate solution is {x1}")
                break

            # next state
            x0 = x1
        
            # max step limit reached
            if step == max_steps-1:
                print(f"Max step limit reached: x0 = {x0} and f(x0) = {f(x0)}, choose a closer point")
                break
        
            list_plots+=[pl+list_plot(list_approx,color='blue',xmin=a-0.1,xmax=b+0.1,ymin=-10,ymax=40)]

    # check for Bisection
    if bisection == True:
        
        # list for midpoints
        root_estimates = []
        
        # time steps
        time = 0
        
        # check if bisection conditions are met
        if f(a) * f(b) > 0:
            print("must be within bounds")
            return
        else:
            while(b-a > error):
                
                # print time steps
                print(time)
                
                # increment time steps
                time += 1
                
                # calculate c such that a <= c <= b
                c = (a + b) / 2
                
                # collect the root estimates
                root_estimates.append(c)
                
                # print each interval and c
                print(f"current interval {a, b} with mid point {float(c)} and f(c) = {float(f(c))} and error of {float((b-a)/2)}")

                if f(c) == 0:
                    print(f"found solution {float(c)}")
                    break

                elif f(a) * f(c) < 0:
                    b = c

                else:
                    a = c

        print(f"we found an exact solution {float(c)}")
     
            
        # animate root estimates
        list_plots += [pl + list_plot(root_estimates,color='green')]
 
    myAnim=animate(list_plots)
    myAnim.show(delay=delay)
```
