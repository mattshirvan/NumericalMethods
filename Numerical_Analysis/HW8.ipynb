- allows the user to enter a set of points and it also has two checkboxes.
- if the first checkbox is checked, then your code will display the line of best fit, along with the value of its sum of squares: ssquares.png   
- if the second checkbox is checked, then your code will display the quadratic function of best fit, along with the value of - - its sum of squares (see notes for this formula)
- regardless of the checkboxes (check or unchecked) your code should always display the points given.

```python
@interact
def regression(points = input_box(default = [(4, 2), (7, 1), (11, 2), (13, 6), (17, 7)]), 
               linear = False, 
               quadratic = False, m1 = input_box(default = 0.25), b1 = input_box(default = 0.4)):
    
    ##############
    # Initialize #
    ##############
    
    # Variables
    var('A B C M')
    
    # Range minimum
    start = min(pt[0] for pt in points)
    
    # Range max
    stop = max(pt[0] for pt in points)
    
    # Lists for x and y points
    X = list(zip(*points))[0]; Y = list(zip(*points))[1]
    
    # Plot points
    pl = list_plot(points, size=50, color='green')
    
    # loss
    loss = loss2 = 0
    
    # Summations
    n = len(points)
    x_sum = sum(X)
    y_sum = sum(Y)
    xy_sum = sum(list(map(lambda x, y: x * y, X, Y)))
    x2_sum = sum(list(map(lambda x: x^2, X)))
    x2y_sum = sum(list(map(lambda x, y: x^2 * y, X, Y)))
    x3_sum = sum(list(map(lambda x: x^3, X)))
    x4_sum = sum(list(map(lambda x: x^4, X)))
    
    # Line of best fit
    if linear:
        
        # Solve equations
        sol=solve([M*x2_sum + B*x_sum == xy_sum, M*x_sum + B*len(points) == y_sum],M,B)
        
        # weight
        w = sol[0][0].rhs()
        
        # bias
        b = sol[0][1].rhs()
        
        # Functions
        f(x) = w*x + b; h(x) = m1*x + b1
        
        # Plot functions
        pl += plot(f(x), start, stop); pl += plot(h(x), start, stop, legend_label="Second Line", color="purple")
    
        # Iterate through points
        for i in range(len(points)):

            # Plot line of best fit
            pl += line([(X[i], Y[i]), (X[i], f(X[i]))], color='red')

            # Plot distance points
            pl += list_plot([(X[i], f(X[i]))], size=50, color='red')

            # Add cost to loss
            loss += (Y[i] - f(X[i]))^2
            
            # Plot second line of best fit
            pl += line([(X[i], Y[i]), (X[i], h(X[i]))], color='orange')

            # Plot second distance points
            pl += list_plot([(X[i], h(X[i]))], size=50, color='orange')

            # Add cost to loss
            loss2 += (Y[i] - h(X[i]))^2
            
        # Show loss
        show(f"Sum of squares line one: {loss.n()}")
        show(f"Sum of squares line two: {loss2.n()}")
            
    # Quadratic
    if quadratic:
        
        # Solve equations
        sol = solve([A*n + B*x_sum + C*x2_sum - y_sum == 0,
                     A*x_sum + B*x2_sum + C*x3_sum - xy_sum == 0,
                     A*x2_sum + B*x3_sum + C*x4_sum - x2y_sum == 0], A,B,C)
        
        # Coefficients and constant in quadratic
        a = sol[0][0].rhs()
        b = sol[0][1].rhs()
        c = sol[0][2].rhs()
        
        # Function
        g(x) = a + b*x + c*x^2
        
        # Plot function
        pl += plot(g(x), start, stop)
    
        # Iterate through points
        for i in range(len(points)):

            # Plot line of best fit
            pl += line([(X[i], Y[i]), (X[i], g(X[i]))], color='red')

            # Plot distance points
            pl += list_plot([(X[i], g(X[i]))], size=50, color='red')

            # Add to cost to loss
            loss += (Y[i] - g(X[i]))^2
        
        # Show loss
        show(f"Sum of squares: {loss.n()}")
    
    # Show plot
    pl.show()
    
    
```
